### Correlations:

```python
#Naming Columns
df.columns
```

```python
#getting rid of text only columns
sub = df[['Total Crime 2021', 'High School Finished Year 12', 'High School Finished Year 10', 'High School Finished Year 8 or below', 'Median Age','Median Mortage Payments', 'Median Family Income Weekly', 'Median Rent Weekly', 'Average Household Size', 'Married', 'Seperated', 'Divorced', 'Widowed', 'Never Married']]
```

```python
sub
```

```python
# Create a correlation matrix using pandas

sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(10, 10)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=10, rotation=90) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=8) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

The primary data that I wanted to see was how different factors affect the total crime rate. It seems that age has no effect at all with the crime rate, with a correlation factor of 0.093. Strangely enough, it seems that there is the most correlation between when people finished high school and total crime. Median wage doesn't seem to have any effect on the crime rate either. There doesn't really seem to be a strong correlation from any of my data points.

### Categorical Plots:

```python
# Creating a histogram

sns.histplot(df['Total Crime 2021'], bins = 70, kde = True)
```

```python
df.loc[df['Total Crime 2021'] < 10000, 'Crime Rate'] = 'Low Crime Area'
```

```python
df.loc[(df['Total Crime 2021'] >= 10000) & (df['Total Crime 2021'] < 35000), 'Crime Rate'] = 'Middle Crime Area'
```

```python
df.loc[df['Total Crime 2021'] >= 35000, 'Crime Rate'] = 'High Crime Area'
```

```python
df['Crime Rate'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the categories created above

sns.set(style="ticks")
g = sns.catplot(x="Median Mortage Payments", y="Median Family Income Weekly", hue="Crime Rate", data=df)
```

```python
# fig = g.get_figure()
g.savefig("out.png") 

# Saving image in the working directory. 
```

You can see that the 2 high crime areas are both in places where the median family income was below 2500. Most of the middle crime areas also appear in places where median mortage payments were also below 2500, but there are a couple above. Low crime rates are represented in all aspects. 

#### Does the median income of the area affect the crime rate?

In 2021 the two highest crime areas were in places where the median family income was below 2500 a week, so the higher crime areas are in places where there is a lower income. However, there is a lot of low crime areas in places that have a lower weekly median, so it doesn't really seem to have an actual affet on the crime rate.
